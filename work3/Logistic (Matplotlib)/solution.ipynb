```python
import numpy as np
import matplotlib.pyplot as plt
from math import e


def main():
    x = np.arange(-10, 10, 0.1, dtype=float)
    L = 1
    k = 1
    x0 = 1
    y = calc(x, L, k, x0)
    plt.plot(x, y)
    plt.plot(x0, L/2, marker='o')
    plt.show()


def calc(x, L, k, x0):
    return L/(1+e**(-k*(x-x0)))


if __name__ == '__main__':
    main()
```

```python
那么我们现在生成了一个基本的Logistic函数，横坐标极限分度值为0.1。
```

```python
为了直观显示，我们将更改过参数后的图像生成在一个图里。
```

```python
import numpy as np
import matplotlib.pyplot as plt
from math import e


def main():
    x = np.arange(-10, 10, 0.1, dtype=float)
    L = 1
    k = 1
    x0 = 1
    y = calc(x, L, k, x0)
    plt.plot(x, y)
    plt.plot(x0, L/2, marker='o')
    L = 2
    y = calc(x, L, k, x0)
    plt.plot(x, y, color='r')
    L = 3
    y = calc(x, L, k, x0)
    plt.plot(x, y, color='r')
    L = 1
    k = 2
    y = calc(x, L, k, x0)
    plt.plot(x, y, color='g')
    k = 3
    y = calc(x, L, k, x0)
    plt.plot(x, y, color='g')
    k = 1
    x0 = 2
    y = calc(x, L, k, x0)
    plt.plot(x, y, color='b')
    x0 = 3
    y = calc(x, L, k, x0)
    plt.plot(x, y, color='b')
    plt.show()


def calc(x, L, k, x0):
    return L/(1+e**(-k*(x-x0)))


if __name__ == '__main__':
    main()
```

```python
由图可见，L值的更改直接影响了函数的纵度，而k值的更改直接影响了函数的增长率但不更改中心点，而x0值的更改只会影响函数的横坐标位置。
```
